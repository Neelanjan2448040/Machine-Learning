## Data Preprocessing:

### Dropping not relevant features:

```python
#Remove Id, Street, Utilities, Heating
df_hp.drop(['Id','Street','Utilities','Heating'],axis=1,inplace=True)
```

### Missing Data:

```python
#Checking for missing values
missing_values=df_hp.isnull().sum()
#Calculating missing percentage
missing_percentage=(missing_values/len(df_hp))*100
missing_data=pd.DataFrame({'Missing Values': missing_values, 'Percentage': missing_percentage})
#Filter to display only columns with missing values
missing_data=missing_data[missing_data['Missing Values']>0]
#Display missing values for allcolumns
print(missing_data)
```

```python
# Drop columns with more than 80% missing values
df_hp.drop(columns=['Alley', 'MiscFeature', 'PoolQC', 'Fence'], inplace=True)
```

```python
# Remove records with missing values in specified columns
df_hp = df_hp.dropna(subset=['Electrical', 'MasVnrType', 'MasVnrArea'])
```

### Imputation for numerical features:

```python
# Impute numerical missing values
df_hp['LotFrontage'].fillna(df_hp['LotFrontage'].median(), inplace=True)
df_hp['GarageYrBlt'].fillna(0, inplace=True)
```

### Imputation for categorical features:

```python
# Impute categorical missing values
garage_cols = ['GarageType', 'GarageFinish', 'GarageQual', 'GarageCond']
df_hp[garage_cols] = df_hp[garage_cols].fillna('None')

bsmt_cat_cols = ['BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2']
df_hp[bsmt_cat_cols] = df_hp[bsmt_cat_cols].fillna('None')

df_hp['FireplaceQu'].fillna('None', inplace=True)

# Verify missing values handled
print("Total missing values left:", df_hp.isnull().sum().sum())
```

### Correlation with SalePrice:

```python
# Select only numeric columns for the correlation matrix
numeric_df = df_hp.select_dtypes(include=['int64', 'float64'])

# Calculate the correlation matrix
corr_matrix = numeric_df.corr()

# Get absolute correlations with SalePrice and sort
top_corr_features = corr_matrix['SalePrice'].abs().sort_values(ascending=False)

# Print the most correlated features with SalePrice
print(top_corr_features)
```

### Takes high correlation values with respect to SalePrice from above

```python
threshold=0.5
high_corr_values=[]
for feature in corr_matrix:
    high_corr_features=corr_matrix.index[(corr_matrix[feature].abs()>threshold)]
    high_corr_values.extend(high_corr_features)
high_corr_values=list(set(high_corr_values))
high_corr_values
```

```python
# Select top 10 correlated features
top_corr_features = top_corr_features.sort_values(ascending=False).head(10)
plt.figure(figsize=(10,5))
sb.barplot(x=top_corr_features.index, y=top_corr_features.values)
plt.xticks(rotation=45)
plt.title("Top 10 features correlated with SalesPrice")
plt.ylabel("Correlation")
plt.show()
```
